````python
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta
import json

# Configuration
API_BASE_URL = 'http://localhost:8000/api'
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette('husl')
````

## 1. Test Backend Health

````python
# Check if backend is running
try:
    response = requests.get('http://localhost:8000/')
    print('✅ Backend Status:', response.json())
except Exception as e:
    print('❌ Backend not available:', str(e))
    print('\nPlease start the backend first:')
    print('cd backend && python main.py')
````

## 2. Inventory Management

````python
# Get all inventory
response = requests.get(f'{API_BASE_URL}/inventory')
inventory = response.json()

# Convert to DataFrame
df_inventory = pd.DataFrame(inventory)
print(f'Total Inventory Items: {len(df_inventory)}')
print('\nStock Status Distribution:')
print(df_inventory['stock_status'].value_counts())

df_inventory.head(10)
````

````python
# Visualize inventory distribution
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Stock status distribution
df_inventory['stock_status'].value_counts().plot(kind='bar', ax=axes[0,0], color=['green', 'orange', 'red'])
axes[0,0].set_title('Stock Status Distribution')
axes[0,0].set_ylabel('Count')

# Lens type distribution
df_inventory['lens_type'].value_counts().plot(kind='pie', ax=axes[0,1], autopct='%1.1f%%')
axes[0,1].set_title('Lens Type Distribution')

# Quantity distribution
df_inventory['quantity'].hist(bins=20, ax=axes[1,0], edgecolor='black')
axes[1,0].set_title('Quantity Distribution')
axes[1,0].set_xlabel('Quantity')
axes[1,0].set_ylabel('Frequency')

# Power sphere vs cylinder
scatter = axes[1,1].scatter(df_inventory['power_sphere'], df_inventory['power_cylinder'], 
                           c=df_inventory['quantity'], cmap='viridis', s=100, alpha=0.6)
axes[1,1].set_title('Prescription Power Distribution')
axes[1,1].set_xlabel('Power Sphere')
axes[1,1].set_ylabel('Power Cylinder')
plt.colorbar(scatter, ax=axes[1,1], label='Quantity')

plt.tight_layout()
plt.show()
````

## 3. Order Management

````python
# Create a test order
new_order = {
    'customer_name': 'John Doe',
    'lens_type': 'Progressive',
    'power_sphere': -2.5,
    'power_cylinder': -1.0,
    'location': 'Mumbai',
    'sla_hours': 48,
    'operator_name': 'Demo Operator'
}

response = requests.post(f'{API_BASE_URL}/orders', json=new_order)
created_order = response.json()

print('✅ Order Created Successfully!')
print(f"Order ID: {created_order['order']['id']}")
print(f"Stock Status: {created_order['stock_status']}")
print(f"Breach Probability: {created_order['order']['breach_probability']}%")
print(f"Risk Level: {created_order['order']['risk_level']}")
````

````python
# Get all orders
response = requests.get(f'{API_BASE_URL}/orders')
orders = response.json()

df_orders = pd.DataFrame(orders)
print(f'Total Orders: {len(df_orders)}')

if len(df_orders) > 0:
    print('\nRisk Level Distribution:')
    print(df_orders['risk_level'].value_counts())
    print('\nCurrent Stage Distribution:')
    print(df_orders['current_stage'].value_counts())
    
    df_orders[['id', 'customer_name', 'current_stage', 'breach_probability', 'risk_level']].head()
````

## 4. ML Model Analysis

````python
# Get ML model metrics
response = requests.get(f'{API_BASE_URL}/model-metrics')
metrics = response.json()

print('ML Model Performance Metrics:')
print(f"Accuracy: {metrics['accuracy']}%")
print(f"Precision: {metrics['precision']}%")
print(f"Recall: {metrics['recall']}%")
print(f"F1 Score: {metrics['f1_score']}%")
print(f"\nTotal Training Samples: {metrics['total_samples']}")

# Feature importance
feature_importance = pd.DataFrame(metrics['feature_importance'])
print('\nFeature Importance:')
feature_importance
````

````python
# Visualize model performance and feature importance
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Model metrics
model_metrics = ['accuracy', 'precision', 'recall', 'f1_score']
model_values = [metrics[m] for m in model_metrics]
axes[0].bar(model_metrics, model_values, color=['#4CAF50', '#2196F3', '#FF9800', '#9C27B0'])
axes[0].set_title('ML Model Performance Metrics', fontsize=14, fontweight='bold')
axes[0].set_ylabel('Score (%)')
axes[0].set_ylim([0, 100])
for i, v in enumerate(model_values):
    axes[0].text(i, v + 2, f'{v}%', ha='center', fontweight='bold')

# Feature importance
feature_importance_sorted = feature_importance.sort_values('importance', ascending=True)
axes[1].barh(feature_importance_sorted['feature'], feature_importance_sorted['importance'], 
             color=sns.color_palette('viridis', len(feature_importance)))
axes[1].set_title('Feature Importance Rankings', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Importance (%)')

plt.tight_layout()
plt.show()
````

## 5. Dashboard Metrics Analysis

````python
# Get dashboard metrics
response = requests.get(f'{API_BASE_URL}/dashboard/metrics')
dashboard = response.json()

print('📊 Dashboard Metrics Summary')
print('=' * 50)
print(f"Total Active Orders: {dashboard['total_active']}")
print(f"Breached Orders: {dashboard['breached']}")
print(f"High Risk Orders: {dashboard['high_risk']}")
print(f"Medium Risk Orders: {dashboard['medium_risk']}")
print(f"Average TAT: {dashboard['average_tat_hours']:.1f} hours")
print(f"\nLocation Distribution: {dashboard['locations_distribution']}")
print(f"Lens Type Distribution: {dashboard['lens_distribution']}")
````

````python
# Visualize dashboard metrics
fig = plt.figure(figsize=(16, 10))
gs = fig.add_gridspec(3, 3, hspace=0.3, wspace=0.3)

# Risk distribution
ax1 = fig.add_subplot(gs[0, :])
risk_categories = ['High Risk', 'Medium Risk', 'Low Risk']
low_risk = dashboard['total_active'] - dashboard['high_risk'] - dashboard['medium_risk']
risk_values = [dashboard['high_risk'], dashboard['medium_risk'], low_risk]
colors = ['#f44336', '#ff9800', '#4caf50']
ax1.bar(risk_categories, risk_values, color=colors, edgecolor='black', linewidth=2)
ax1.set_title('Order Risk Distribution', fontsize=16, fontweight='bold')
ax1.set_ylabel('Number of Orders')
for i, v in enumerate(risk_values):
    ax1.text(i, v + 0.5, str(v), ha='center', fontweight='bold', fontsize=12)

# Location distribution
if dashboard['locations_distribution']:
    ax2 = fig.add_subplot(gs[1, 0])
    locations = list(dashboard['locations_distribution'].keys())
    location_counts = list(dashboard['locations_distribution'].values())
    ax2.pie(location_counts, labels=locations, autopct='%1.1f%%', startangle=90)
    ax2.set_title('Orders by Location', fontweight='bold')

# Lens type distribution
if dashboard['lens_distribution']:
    ax3 = fig.add_subplot(gs[1, 1])
    lens_types = list(dashboard['lens_distribution'].keys())
    lens_counts = list(dashboard['lens_distribution'].values())
    ax3.pie(lens_counts, labels=lens_types, autopct='%1.1f%%', startangle=90)
    ax3.set_title('Orders by Lens Type', fontweight='bold')

# KPI Summary
ax4 = fig.add_subplot(gs[1, 2])
ax4.axis('off')
kpi_text = f"""
KEY PERFORMANCE INDICATORS

Total Active Orders: {dashboard['total_active']}
Breached Orders: {dashboard['breached']}
Breach Rate: {(dashboard['breached']/max(dashboard['total_active'],1)*100):.1f}%

High Risk: {dashboard['high_risk']}
Medium Risk: {dashboard['medium_risk']}
Low Risk: {low_risk}

Average TAT: {dashboard['average_tat_hours']:.1f}h
Target SLA: 48.0h
"""
ax4.text(0.1, 0.5, kpi_text, fontsize=11, family='monospace', 
         bbox=dict(boxstyle='round', facecolor='lightblue', alpha=0.5))

# ML Model metrics
ax5 = fig.add_subplot(gs[2, :])
ml_metrics = dashboard['ml_classifier_state']
metric_names = ['Accuracy', 'Precision', 'Recall', 'F1 Score']
metric_values = [ml_metrics['accuracy'], ml_metrics['precision'], 
                ml_metrics['recall'], ml_metrics['f1_score']]
ax5.plot(metric_names, metric_values, marker='o', linewidth=3, markersize=10)
ax5.set_title('ML Model Performance Metrics', fontsize=14, fontweight='bold')
ax5.set_ylabel('Score (%)')
ax5.set_ylim([85, 100])
ax5.grid(True, alpha=0.3)
for i, v in enumerate(metric_values):
    ax5.text(i, v + 0.5, f'{v}%', ha='center', fontweight='bold')

plt.suptitle('Eluno AI OMS - Dashboard Overview', fontsize=18, fontweight='bold', y=0.98)
plt.show()
````

## 6. Alert System Testing

````python
# Get all alerts
response = requests.get(f'{API_BASE_URL}/alerts')
alerts = response.json()

if alerts:
    df_alerts = pd.DataFrame(alerts)
    print(f'Total Alerts Generated: {len(df_alerts)}')
    print('\nAlert Channel Distribution:')
    print(df_alerts['channel'].value_counts())
    print('\nRecent Alerts:')
    df_alerts[['id', 'order_id', 'customer_name', 'risk_score', 'channel', 'sent_at']].head(10)
else:
    print('No alerts generated yet')
````

## 7. AI Recommendation System Test

````python
# Get a high-risk order for recommendation
response = requests.get(f'{API_BASE_URL}/orders')
orders = response.json()

if orders:
    # Find first order (or create one for testing)
    test_order_id = orders[0]['id']
    
    # Request AI recommendation
    response = requests.post(
        f'{API_BASE_URL}/recommendations',
        json={'order_id': test_order_id}
    )
    
    if response.status_code == 200:
        recommendation = response.json()['recommendation']
        print(f'🤖 AI Recommendation for Order {test_order_id}')
        print('=' * 80)
        print(recommendation)
    else:
        print(f'Failed to get recommendation: {response.status_code}')
else:
    print('No orders available for recommendation testing')
````

## 8. Breach Probability Prediction Analysis

````python
# Analyze breach probability across different scenarios
if len(df_orders) > 0:
    fig, axes = plt.subplots(2, 2, figsize=(15, 10))
    
    # Breach probability by current stage
    stage_risk = df_orders.groupby('current_stage')['breach_probability'].mean().sort_values(ascending=False)
    stage_risk.plot(kind='barh', ax=axes[0,0], color='coral')
    axes[0,0].set_title('Average Breach Probability by Stage')
    axes[0,0].set_xlabel('Breach Probability (%)')
    
    # Breach probability by lens type
    lens_risk = df_orders.groupby('lens_type')['breach_probability'].mean()
    lens_risk.plot(kind='bar', ax=axes[0,1], color='skyblue')
    axes[0,1].set_title('Average Breach Probability by Lens Type')
    axes[0,1].set_ylabel('Breach Probability (%)')
    axes[0,1].set_xticklabels(axes[0,1].get_xticklabels(), rotation=45)
    
    # Risk level distribution
    risk_dist = df_orders['risk_level'].value_counts()
    risk_dist.plot(kind='pie', ax=axes[1,0], autopct='%1.1f%%', 
                   colors=['green', 'orange', 'red'])
    axes[1,0].set_title('Risk Level Distribution')
    axes[1,0].set_ylabel('')
    
    # Stock status impact on breach probability
    stock_risk = df_orders.groupby('stock_status')['breach_probability'].mean()
    stock_risk.plot(kind='bar', ax=axes[1,1], color='lightgreen')
    axes[1,1].set_title('Breach Probability by Stock Status')
    axes[1,1].set_ylabel('Breach Probability (%)')
    axes[1,1].set_xticklabels(axes[1,1].get_xticklabels(), rotation=45)
    
    plt.tight_layout()
    plt.show()
````
